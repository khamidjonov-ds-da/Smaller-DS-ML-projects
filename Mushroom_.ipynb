Import essential libs

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import
```

Load data

```python
mushroom_data = pd.read_csv('/content/mushrooms.csv')
mushroom_data.sample(5)
```

```python
mushroom_data.info()
```

```python
mushroom_data.corr()
```

Converting into numerical data

```python
from sklearn.preprocessing import LabelEncoder

labelEncoder = LabelEncoder()

# The data is categorical
for i in mushroom_data.columns:
  mushroom_data[i] = labelEncoder.fit_transform(mushroom_data[i])

mushroom_data.sample(5)
```

Train and test split

```python
from sklearn.model_selection import train_test_split

x_data = mushroom_data.drop("class", axis=1)
y_data = mushroom_data["class"]

x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2, random_state=2)
```

Training model with Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

logReg = LogisticRegression(max_iter=500)
logReg.fit(x_train, y_train)
predictions = logReg.predict(x_test)
```

Accuracy score

```python
logReg.score(x_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, predictions)
```
